```python
import geopandas as gpd
from shapely.geometry import MultiPolygon, Polygon
import ee

# Ensure geometries are reprojected to WGS84
ventenata_montana = ventenata_montana.to_crs("EPSG:4326")

# Replace NaN values with an empty string or None
ventenata_montana = ventenata_montana.fillna("")

def simplify_geometry(gdf):
    """
    Converts a GeoDataFrame into an Earth Engine FeatureCollection,
    handling MultiPolygon geometries by splitting them into individual Polygons.
    """
    features = []
    for _, row in gdf.iterrows():
        geometry = row.geometry

        # Convert all row data to a dictionary and handle NaN/None
        properties = {k: (v if v is not None else "") for k, v in row.drop("geometry").to_dict().items()}

        # Check if the geometry is a MultiPolygon
        if isinstance(geometry, MultiPolygon):
            for polygon in geometry.geoms:
                coords = [[list(point) for point in polygon.exterior.coords]]
                features.append(
                    ee.Feature(
                        ee.Geometry.Polygon(coords),
                        properties
                    )
                )
        elif isinstance(geometry, Polygon):
            coords = [[list(point) for point in geometry.exterior.coords]]
            features.append(
                ee.Feature(
                    ee.Geometry.Polygon(coords),
                    properties
                )
            )
        else:
            print(f"Unsupported geometry type: {type(geometry)}")

    return ee.FeatureCollection(features)

# Convert the GeoDataFrame to an Earth Engine FeatureCollection
try:
    gee_polygons = simplify_geometry(ventenata_montana)

    # Print a single feature for debugging
    print(gee_polygons.limit(1).getInfo())
except Exception as e:
    print(f"Error: {e}")
```

### Calculate area of each polygon

```python
# Add an area property to each feature
def calculate_area(feature):
    area = feature.geometry().area().divide(1e6)  # Convert to square kilometers
    return feature.set('calculated_area_km2', area)

# Apply the function to each feature
area_calculated = gee_polygons.map(calculate_area)

# Print results to verify
print(area_calculated.getInfo())
```

### Remove Polygons with Area Less than Threshold

```python
# Filter features with area > 0.01 km (10,000 m)
filtered_polygons = area_calculated.filter(ee.Filter.gt('calculated_area_km2', 0.01))
```

### Buffer Polygons by 100 Meters

```python
# Buffer each polygon
def add_buffer(feature):
    buffered_geometry = feature.geometry().buffer(500)  # 500 meters
    return feature.setGeometry(buffered_geometry)

buffered_polygons = area_calculated.map(add_buffer)

# Print results to verify
print(buffered_polygons.getInfo())
```

### CNN Model

	1.	Load and Preprocess Imagery
Load a relevant satellite image collection (e.g., Sentinel-2, Landsat) and preprocess it to ensure the data is suitable (e.g., cloud masking, selecting bands or indices).

```python
# Import the required library
import ee
ee.Initialize()

# Load Sentinel-2 imagery
sentinel = ee.ImageCollection("COPERNICUS/S2") \
    .filterDate('2018-01-01', '2023-12-31') \
    .filterBounds(buffered_polygons) \
    .map(lambda img: img.clip(buffered_polygons))  # Clip imagery to the polygons

# Select relevant bands (e.g., Red, Green, Blue, and NDVI)
bands = ['B4', 'B3', 'B2', 'NDVI']
def add_ndvi(image):
    ndvi = image.normalizedDifference(['B8', 'B4']).rename('NDVI')
    return image.addBands(ndvi)

sentinel = sentinel.map(add_ndvi).select(bands)
```

	2.	Clip Imagery by Polygons
Use each polygon from your FeatureCollection to clip the imagery and create training data.

```python
# Function to extract band values for a single polygon from a single image
def compute_stats_for_feature(image):
    def reduce_region(feature):
        # Compute mean band values for the feature
        stats = image.reduceRegion(
            reducer=ee.Reducer.mean(),
            geometry=feature.geometry(),
            scale=10,  # Adjust scale to match image resolution
            maxPixels=1e13
        )
        # Set the computed stats as properties on the feature
        return feature.set(stats)
    return buffered_polygons.map(reduce_region)

# Apply the function to all images in the collection
def extract_band_values_from_image_collection(image):
    return compute_stats_for_feature(image)

# Map over the ImageCollection to extract training data for each image
mapped_collections = sentinel.map(extract_band_values_from_image_collection)

# Combine all FeatureCollections from each image into a single FeatureCollection
def merge_collections(image_fc, previous_fc):
    return ee.FeatureCollection(previous_fc).merge(ee.FeatureCollection(image_fc))

flattened_results = ee.FeatureCollection(mapped_collections.iterate(
    merge_collections,
    ee.FeatureCollection([])
))

# Display the resulting training data
print(flattened_results.getInfo())

# Optionally export the training data as a CSV
task = ee.batch.Export.table.toDrive(
    collection=flattened_results,
    description='TrainingDataExport',
    fileFormat='CSV'
)
task.start()
```
